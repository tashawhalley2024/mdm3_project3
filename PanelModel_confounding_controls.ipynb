```python
#upload file containing data to be added

#Adding Corruption Index
VDem_datafile_path = '/content/drive/MyDrive/Panel_Model_Control_Data/Corruption/V-Dem-CY-Full+Others-v16.csv'
VDem_datafile = pd.read_csv(VDem_datafile_path)
VDem_df = pd.DataFrame(VDem_datafile)

with open(VDem_datafile_path) as VDem_datafile:
  VDem_reader = csv.reader(VDem_datafile, delimiter=',')
  VDem_headers = next(VDem_reader)

print ('Column Names', VDem_headers)

```

```python
corruption_cols = ['country_name', 'country_text_id', 'country_id', 'year', 'v2x_corr', 'v2x_execorr','v2x_pubcorr','v2lgcrrpt', 'v2jucorrdc']

required_corruption_dataset = VDem_df[corruption_cols]

#years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

corruption_prev_years_rows = required_corruption_dataset[required_corruption_dataset['year'] < 2013].index

required_corruption_dataset.drop(corruption_prev_years_rows, inplace=True)

print (required_corruption_dataset)
```

```python
#find whether their ID's (iso3's and country_text_id) are equivalent to one another

main_iso3_list = list(set(main_df['iso3']))
main_iso3_list.sort()
required_VDem_iso3_list = list(set(required_corruption_dataset['country_text_id']))
required_VDem_iso3_list.sort()


print (len(main_iso3_list))
print (main_iso3_list)
print (required_VDem_iso3_list)
print (len(required_VDem_iso3_list))

unmatched_iso3s_tomain = []
unmatched_iso3s_tocorr = []

countries_not_in_main  = []

for country in required_VDem_iso3_list:
  if country not in main_iso3_list:
    if country not in unmatched_iso3s_tomain:
      unmatched_iso3s_tomain.append(country)
      countries_not_in_main.append([country, required_corruption_dataset.loc[required_corruption_dataset['country_text_id'] == country, 'country_name'].iloc[0]])

print(countries_not_in_main)

countries_not_in_corruption_data = []

for country in main_iso3_list:
  if country not in required_VDem_iso3_list:
    unmatched_iso3s_tocorr.append(country)
    countries_not_in_corruption_data.append([country, main_df.loc[main_df['iso3'] == country, 'country'].iloc[0]])

print (countries_not_in_corruption_data)

print (len(unmatched_iso3s_tomain), unmatched_iso3s_tomain)
print ( len(unmatched_iso3s_tocorr), unmatched_iso3s_tocorr)
```

```python
#check if any countries said not in corruption were just under a different code/ iso3
for code, name in countries_not_in_corruption_data:
  if name in required_corruption_dataset['country_name'].values:
    print (name)


for code, name in countries_not_in_main:
  if name in main_df['country'].values:
    print (name)

#No overlaps - just missing countries/ data
```

```python
#append onto the file / create a new file with the same first three columns - #iso3, #country, #year
final_added_corr_df = main_df.merge(required_corruption_dataset, left_on=['iso3', 'year'], right_on=['country_text_id', 'year'], how='left')
final_added_corr_df.drop(columns=['country_text_id','country_name', 'country_id'], inplace=True)
print (final_added_corr_df)
```

```python
democracy_cols = ['country_name', 'country_text_id', 'country_id', 'year', 'v2x_libdem']

required_democracy_dataset = VDem_df[democracy_cols]

democracy_prev_years_rows = required_democracy_dataset[required_democracy_dataset['year'] < 2013].index

required_democracy_dataset.drop(democracy_prev_years_rows, inplace=True)

print (required_democracy_dataset)
```

```python
#append onto the file / create a new file with the same first three columns - #iso3, #country, #year
final_added_democracy_df = final_added_corr_df.merge(required_democracy_dataset, left_on=['iso3', 'year'], right_on=['country_text_id', 'year'], how='left')
final_added_democracy_df.drop(columns=['country_text_id','country_name', 'country_id'], inplace=True)
print (final_added_democracy_df)
```

```python
#Adding Education Index
education_datafile_path = '/content/drive/MyDrive/Panel_Model_Control_Data/Education/hdr_avg_education_data.csv'
education_datafile = pd.read_csv(education_datafile_path)
education_df = pd.DataFrame(education_datafile)

with open(education_datafile_path) as education_datafile:
  education_reader = csv.reader(education_datafile, delimiter=',')
  education_headers = next(education_reader)

print ('Education Data Column Names', education_headers)

print (education_df)
```

```python
required_education_cols =['countryIsoCode', 'country', 'year', 'value']

required_education_dataset = education_df[required_education_cols]

print (required_education_dataset)

```

```python
#find whether their ID's (iso3's and countryIsoCode) are equivalent to one another

required_Edu_iso3_list = list(set(required_education_dataset['countryIsoCode']))
required_Edu_iso3_list.sort()


print (len(main_iso3_list))
print (main_iso3_list)
print (required_Edu_iso3_list)
print (len(required_Edu_iso3_list))

unmatched_iso3s_tomain_ed = []
unmatched_iso3s_toEdu_main = []

countries_not_in_main_frm_edu  = []

for country in required_Edu_iso3_list:
  if country not in main_iso3_list:
    if country not in unmatched_iso3s_tomain_ed:
      unmatched_iso3s_tomain_ed.append(country)
      countries_not_in_main_frm_edu.append([country, required_education_dataset.loc[required_education_dataset['countryIsoCode'] == country, 'country'].iloc[0]])

print(countries_not_in_main_frm_edu)

countries_not_in_edu_data = []

for country in main_iso3_list:
  if country not in required_Edu_iso3_list:
    unmatched_iso3s_toEdu_main.append(country)
    countries_not_in_edu_data.append([country, main_df.loc[main_df['iso3'] == country, 'country'].iloc[0]])

print (countries_not_in_edu_data)

print (len(unmatched_iso3s_tomain_ed), unmatched_iso3s_tomain_ed)
print ( len(unmatched_iso3s_toEdu_main), unmatched_iso3s_toEdu_main)
```

```python
#check if any countries said not in education/ main were just under a different code/ iso3
for code, name in countries_not_in_edu_data:
  if name in required_education_dataset['country'].values:
    list1.append(name)
    print (name)


for code, name in countries_not_in_main_frm_edu:
  if name in main_df['country'].values:
    print (name)

#No overlaps - just missing countries/ data
```

```python
#append onto the file / create a new file with the same first three columns - #iso3, #country, #year
final_added_edu_df = final_added_democracy_df.merge(required_education_dataset, left_on=['iso3', 'year'], right_on=['countryIsoCode', 'year'], how='left')
final_added_edu_df.drop(columns=['countryIsoCode', 'country_y'], inplace=True)
final_added_edu_df.rename(columns={'value': 'education_index'}, inplace=True)
print (final_added_edu_df)
```

```python
#add rurality dataset

rurality_datafile_path = '/content/drive/MyDrive/Panel_Model_Control_Data/Rurality/Rurality_Data.csv'
rurality_datafile = pd.read_csv(rurality_datafile_path, skiprows = 3, header = 0 )
rurality_df_1 = pd.DataFrame(rurality_datafile)

with open(rurality_datafile_path) as rurality_datafile:
  rurality_reader = csv.reader(rurality_datafile, delimiter=',')
  rurality_headers = next(rurality_reader)

#print ('Rurality Data Column Names', rurality_headers)

rurality_df_2 = rurality_df_1.iloc[2:].reset_index(drop=True)

# Drop year columns before 2013
year_cols_to_drop = [col for col in rurality_df_2.columns
                     if str(col).replace('.0','').strip().isdigit()
                     and not (2013 <= int(str(col).replace('.0','').strip()) <= 2023)]

rurality_df_3 = rurality_df_2.drop(columns=year_cols_to_drop)

# Rename remaining year columns to clean integers
rurality_df = rurality_df_3.rename(columns={col: int(float(col))
                                           for col in rurality_df_3.columns
                                           if str(col).replace('.0','').strip().isdigit()})



print (rurality_df)
```

```python
print(rurality_df.columns.tolist())
```

```python
required_rural_cols = ['Country Name', 'Country Code', 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

required_rural_dataset_1 = rurality_df[required_rural_cols]

print(required_rural_dataset_1)
```

```python
required_rurality_dataset = pd.melt(required_rural_dataset_1, id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Rurality')

required_rurality_dataset['Year'] = required_rurality_dataset['Year'].astype(int)

print (required_rurality_dataset)
```

```python
#find whether their ID's (iso3's and Country Code) are equivalent to one another

required_rurality_iso3_list = list(set(required_rurality_dataset['Country Code']))
required_rurality_iso3_list.sort()


print (len(main_iso3_list))
print (main_iso3_list)
print (required_rurality_iso3_list)
print (len(required_rurality_iso3_list))

unmatched_iso3s_tomain_rural = []
unmatched_iso3s_torural_main = []

countries_not_in_main_frm_rural  = []

for country in required_rurality_iso3_list:
  if country not in main_iso3_list:
    if country not in unmatched_iso3s_tomain_rural:
      unmatched_iso3s_tomain_rural.append(country)
      countries_not_in_main_frm_rural.append([country, required_rurality_dataset.loc[required_rurality_dataset['Country Code'] == country, 'Country Name'].iloc[0]])

print('countries not in main data:', countries_not_in_main_frm_rural)

countries_not_in_rural_data = []

for country in main_iso3_list:
  if country not in required_rurality_iso3_list:
    unmatched_iso3s_torural_main.append(country)
    countries_not_in_rural_data.append([country, main_df.loc[main_df['iso3'] == country, 'country'].iloc[0]])

print ('countries not in rural data:', countries_not_in_rural_data)

print (len(unmatched_iso3s_tomain_rural), unmatched_iso3s_tomain_rural)
print (len(unmatched_iso3s_torural_main), unmatched_iso3s_torural_main)
```

```python
#append onto the file / create a new file with the same first three columns - #iso3, #country, #year
final_added_rurality_df = final_added_edu_df.merge(required_rurality_dataset, left_on=['iso3', 'year'], right_on=['Country Code', 'Year'], how='left')
final_added_rurality_df.drop(columns=['Country Name', 'Country Code'], inplace=True)
print (final_added_rurality_df)
```

```python
#add conflict - Political Instability and Absence of Violence World Bank dataset

WB_conflict_datafile_path = '/content/drive/MyDrive/Panel_Model_Control_Data/Conflict  /WorldBank_PoliticalStability_AbsenceViolence_Data.csv'
WB_conflict_datafile = pd.read_csv(WB_conflict_datafile_path)
WB_conflict_df = pd.DataFrame(WB_conflict_datafile)

with open(WB_conflict_datafile_path) as WB_conflict_datafile:
  WB_conflict_reader = csv.reader(WB_conflict_datafile, delimiter=',')
  WB_conflict_headers = next(WB_conflict_reader)

print ('WB conflict headers', WB_conflict_headers)
print (WB_conflict_df)
```

```python


required_WBconflict_cols = ['Country Name', 'Country Code', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]', '2022 [YR2022]', '2023 [YR2023]']

required_WBconflict_dataset = WB_conflict_df[required_WBconflict_cols]

year_headers_shorten_list = ['2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]', '2022 [YR2022]', '2023 [YR2023]']

required_WBconflict_dataset.rename(columns=lambda col: col[:4] if col in year_headers_shorten_list else col, inplace = True)

print (required_WBconflict_dataset)
```

```python
required_WBconflict_dataset = pd.melt(required_WBconflict_dataset, id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Conflict_Scale')

print (required_WBconflict_dataset)
```

```python
#find which countries are and are not in this dataset and the main dataset  (iso3 and 'Country Code')

# Filter out NaN values before creating the list to avoid TypeError during sorting
required_WBconflict_iso3_list = list(set(required_WBconflict_dataset['Country Code'].dropna()))

print (required_WBconflict_iso3_list)

required_WBconflict_iso3_list.sort()

print (len(main_iso3_list))
print (main_iso3_list)
print (required_WBconflict_iso3_list)
print (len(required_WBconflict_iso3_list))

unmatched_iso3_tomain_WBconflict = []
unmatched_iso3_toWBconflict_main = []

countries_not_in_main_frm_WBconflict  = []

for country in required_WBconflict_iso3_list:
  if country not in main_iso3_list:
    if country not in unmatched_iso3_tomain_WBconflict:
      unmatched_iso3_tomain_WBconflict.append(country)
      countries_not_in_main_frm_WBconflict.append([country, required_WBconflict_dataset.loc[required_WBconflict_dataset['Country Code'] == country, 'Country Name'].iloc[0]])

print ('countries not in main data:', countries_not_in_main_frm_WBconflict)

countries_not_in_WBconflict_data = []

for country in main_iso3_list:
  if country not in required_WBconflict_iso3_list:
    unmatched_iso3_toWBconflict_main.append(country)
    countries_not_in_WBconflict_data.append([country, main_df.loc[main_df['iso3']==country, 'country'].iloc[0]])

print ('countries not in WB conflict data:', countries_not_in_WBconflict_data)

print (len(unmatched_iso3_tomain_WBconflict), unmatched_iso3_tomain_WBconflict)
print (len(unmatched_iso3_toWBconflict_main), unmatched_iso3_toWBconflict_main)
```
